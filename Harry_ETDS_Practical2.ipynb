```python
#Practical 2: Data Integration and Data Transformation for data mining
import pandas as pd
import numpy as np
```

```python
#Samples for datasets for Data integration
data1 = {
    'ID':[1,2,3,4,],
    'Name':['A','B','C','D'],
    'Age':[20,25,30,35]
}
data2 = {
    'ID':[3,4,5,6],
    'Gender':['M','F','M','F'],
    'Salary':[50000,60000,70000,80000]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

##Data Integration

```python
#Tight Coupling (Join datasets on a common key)
harry_tight_coupling = pd.merge(df1,df2,on='ID',how='inner')
print('Tight Coupling Result: \n', harry_tight_coupling)
```

```python
#Loose Coupling (Concatenate Datasets)
harry_loose_coupling = pd.concat([df1.set_index('ID'), df2.set_index('ID')], axis = 1).reset_index()
print("Loose Coupling Result :\n", harry_loose_coupling)
```

##Data Transformation

```python
#Smoothing (Moving Average for Age)
harry_loose_coupling['Smoothed_Age'] = harry_loose_coupling['Age'].rolling(window=2, min_periods=1).mean()
print("Smoothed Age Result :\n", harry_loose_coupling[['ID','Age','Smoothed_Age']])
```

```python
#Aggregation (Summarizing Salary By Gender)
aggregation = harry_loose_coupling.groupby('Gender')['Salary'].sum().reset_index()
print("Aggregation Result :\n", aggregation)
```

```python
#Discretization (Binning Age Into Categories)
bins = [0, 20, 30, 40, 50]
labels = ['Teenager', 'Young Adult', 'Adult', 'Senior']
harry_loose_coupling['Age_Category'] = pd.cut(harry_loose_coupling['Age'], bins=bins, labels=labels)
print("Discretization Result :\n", harry_loose_coupling[['ID','Age','Age_Category']])
```

```python
#Attribute Construction (Creating Age-Salary Ratio)
harry_loose_coupling['Age_Salary_Ratio'] = harry_loose_coupling['Age'] / harry_loose_coupling['Salary']
print("Attribute Construction Result :\n", harry_loose_coupling[['ID','Age','Salary','Age_Salary_Ratio']])
```

```python
#Generalization (Simplifying Salary into Ranges)
salary_bins = [0,60000,80000,100000]
salary_labels = ['Low','Medium','High']
harry_loose_coupling['Salary_Range'] = pd.cut(harry_loose_coupling['Salary'], bins=salary_bins, labels=salary_labels)
print("Generalization Result :\n", harry_loose_coupling[['ID','Salary','Salary_Range']])
```

##Normalization

```python
#Min Max Normalization
harry_loose_coupling['Age_MinMax']=(harry_loose_coupling['Age']-harry_loose_coupling['Age'].min())/(harry_loose_coupling['Age'].max()-harry_loose_coupling['Age'].min())
print("Min Max Normalization Result :\n", harry_loose_coupling[['ID','Age','Age_MinMax']])
```

```python
#Z-Score Normalization
harry_loose_coupling['Age_ZScore']=(harry_loose_coupling['Age']-harry_loose_coupling['Age'].mean())/harry_loose_coupling['Age'].std()
print("\nZ-Score Normalization:\n", harry_loose_coupling[['ID','Age','Age_ZScore']])
```

```python
#Decimal Scaling
scaling_factor = 10**np.ceil(np.log10(harry_loose_coupling['Age'].abs().max()))
harry_loose_coupling['Age_Decimal'] = harry_loose_coupling['Age'] /scaling_factor
print("\nDecimal Scaling :\n", harry_loose_coupling[['ID','Age','Age_Decimal']])
```
