#Practical 1: Data Aggregation and its types with time and spatial aggregation in data mining

```python
import pandas as pd
```

```python
#Step 1: Create A Sample Dataset
data_harry = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South','East','West'],
    'Product': ['A', 'A', 'B', 'C', 'B', 'C','A','D'],
    'Sales': [100, 150, 200, 120, 180, 300, 250, 220],
    'Quantity': [5, 8, 12, 6, 10, 15, 24, 10]
}

df_harry = pd.DataFrame(data_harry)
print(df_harry)
```

```python
#Step 2: Grouping and Aggregation
#i] Aggregating Sales by Region (Sum Aggregation)

sales_by_region_harry = df_harry.groupby('Region')['Sales'].sum()
print(sales_by_region_harry)
```

```python
#Aggregation Sales and Quantity by Product (Mean)
mean_by_product_harry = df_harry.groupby('Product')[['Sales', 'Quantity']].mean()
print(mean_by_product_harry)
```

```python
#Aggregating Count of Sales by region (Count Aggregation)
count_by_region_harry = df_harry.groupby('Region')['Sales'].count()
print(count_by_region_harry)
```

```python
#Custom Aggeregation (Min and Max sales by region)
custom_agg_harry = df_harry.groupby('Region')['Sales'].agg(['min', 'max'])
print(custom_agg_harry)
```

```python
#Step 3: Multi-Level Aggregation
#Aggregating Sales by Region and Product
multi_level_agg_harry = df_harry.groupby(['Region', 'Product'])['Sales'].sum()
print(multi_level_agg_harry)
```

```python
#Step 4: Reset Index for Multi-Level Aggregation
multi_level_agg_harry_reset = multi_level_agg_harry.reset_index()
print(multi_level_agg_harry_reset)
```

#Objective : To understand and implement:

##Time Aggregation: Aggregating data over different time periods.           Spatial Aggregation: Aggregating data by spatial attributes (eg, by region, city)

```python
import pandas as pd
```

```python
#Extend the Dataset with Time and Spatial Data
```

```python
data_harry = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South','East','West'],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'New York', 'Los Angeles','Chicago','Houston'],
    'Product': ['A', 'A', 'B', 'C', 'B', 'C','A','D'],
    'Sales': [100, 150, 200, 120, 180, 300, 250, 220],
    'Quantity': [5, 8, 12, 6, 10, 15, 24, 10],
    'Date': pd.to_datetime(['2024-01-01', '2024-01-10', '2024-02-01', '2024-02-03','2024-06-07','2024-06-15','2024-07-01','2024-07-05'])
}
```

```python
df_harry = pd.DataFrame(data_harry)
print(df_harry)
```

###Time Aggregation

```python
#Step 1: Set date column as Index (optional step)
df_harry.set_index('Date', inplace=True)
```

```python
#Aggregating Sales by month
monthly_sales_harry = df_harry.resample('ME')['Sales'].sum()
print(monthly_sales_harry)
```

```python
#Aggregating Sales by Quarter
quarterly_sales_harry = df_harry.resample('QE')['Sales'].sum()
print(quarterly_sales_harry)
```

```python
#Aggregating sales by year
yearly_sales_harry = df_harry.resample('YE')['Sales'].sum()
print(yearly_sales_harry)
```

```python
#Reset Index to restore original structure
df_harry.reset_index(inplace=True)
```

###Spatial Aggregation

```python
#Aggregating Sales by Region
region_sales_harry = df_harry.groupby('Region')['Sales'].sum()
print(region_sales_harry)
```

```python
#Aggregating Sales by city
city_sales_harry = df_harry.groupby('City')['Sales'].sum()
print(city_sales_harry)
```

```python
#Aggregating Sales by Region and City
sales_by_region_city = df_harry.groupby(['Region', 'City'])['Sales'].sum()
print(sales_by_region_city)
```

```python
#Reseting the index
sales_by_region_city_reset = sales_by_region_city.reset_index()
print(sales_by_region_city_reset)
```

```python
sales_by_region_city_reset.to_csv('sales_by_region_city.csv', index=False)
print("\nSpatial Aggregation data saved to a csv")
```
